```python
mean = data['price_per_sqft'].mean()
std_dev =data['price_per_sqft'].std()
mean,std_dev
lower_bound = mean - 3 * std_dev
upper_bound = mean + 3 * std_dev
data1= data[(data['price_per_sqft']<= upper_bound) & (data['price_per_sqft']>= lower_bound)]
data1
```

```python
new_data=data[(data['price_per_sqft']<=upper_bound) & (data['price_per_sqft']>=lower_bound)]
print('before outliers:',len(data))
print('after outliers:',len(new_data))
print('outliers:', len(data)-len(new_data))
```

```python
# find limits
upper_limit=data['price_per_sqft'].mean()+3*data['price_per_sqft'].std()
lower_limit=data['price_per_sqft'].mean()-3*data['price_per_sqft'].std()
print("upper limit:", upper_limit)
print("lower limit:", lower_limit)
```

```python
# find the outliers
data1=data[(data['price_per_sqft']>upper_limit)| (data['price_per_sqft']<lower_limit)]
data1
```

```python
#trimming or delete the outliers
new_data=data[(data['price_per_sqft']<=upper_limit) & (data['price_per_sqft']>=lower_limit)]
print('before outliers:',len(data))
print('after outliers:',len(new_data))
print('outliers:', len(data)-len(new_data))
```

```python
sns.boxplot(new_data['price_per_sqft'])
```

```python
# capping
new_data=data.copy()
new_data.loc[new_data['price_per_sqft']>upper_limit,'price_per_sqft']=upper_limit
new_data.loc[new_data['price_per_sqft']<lower_limit,'price_per_sqft']=lower_limit
```

```python
#IQR
q1=data['price_per_sqft'].quantile(0.25)
q3=data['price_per_sqft'].quantile(0.75)
iqr=q3-q1
q1,q3,iqr
```

```python
upper_limit=q3+(1.5*iqr)
lower_limit=q1-(1.5*iqr)
upper_limit, lower_limit
```

```python
sns.boxplot(data['price_per_sqft'])
```

```python
data1=data[(data['price_per_sqft']>upper_limit)| (data['price_per_sqft']<lower_limit)]
data1
```

```python
#trimming or delete the outliers
new_data=data[(data['price_per_sqft']<=upper_limit) & (data['price_per_sqft']>=lower_limit)]
print('before outliers:',len(data))
print('after outliers:',len(new_data))
print('outliers:', len(data)-len(new_data))
```

```python
# capping
new_data=data.copy()
new_data.loc[new_data['price_per_sqft']>upper_limit,'price_per_sqft']=upper_limit
new_data.loc[new_data['price_per_sqft']<lower_limit,'price_per_sqft']=lower_limit
```

```python
sns.boxplot(new_data['price_per_sqft'])
```

```python
sns.distplot(new_data['price_per_sqft'])
```

```python
#Q3 Percentile and IQR methods remove more extreme values than the Mean & Std Dev and Z Score methods, 
#compressing the range of data in price_per_sqft significantly.
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=new_data, x='price_per_sqft')
plt.title("Box Plot of Price per Square Foot (After Outlier Removal)")
plt.xlabel("Price per Square Foot")
plt.show()
```

```python
from scipy.stats import skew, kurtosis

plt.figure(figsize=(12, 6))
sns.histplot(new_data['price_per_sqft'], kde=True, bins=30)
plt.title("Histogram of Price per Square Foot (After Outlier Removal)")
plt.xlabel("Price per Square Foot")
plt.ylabel("Frequency")
plt.show()

price_per_sqft_skew = skew(new_data['price_per_sqft'])
price_per_sqft_kurtosis = kurtosis(new_data['price_per_sqft'])

price_per_sqft_skew, price_per_sqft_kurtosis
```
